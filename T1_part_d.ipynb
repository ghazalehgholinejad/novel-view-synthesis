```python
# from google.colab.patches import cv2_imshow
import cv2
import numpy as np

cap = cv2.VideoCapture('E:\_amirkabir/3.mp4')  # Video input

# Define image dimensions for use in bounding boxes
frame_height, frame_width, _ = cap.read()[1].shape

# Set a noise threshold value
noise_threshold = 2000

# Read the first two frames
ret, frame1 = cap.read()
ret, frame2 = cap.read()
```

```python

while ret:
    # Compute the absolute difference between two consecutive frames
    frame_diff = cv2.absdiff(frame1, frame2)

    # Convert the difference to a grayscale image
    gray = cv2.cvtColor(frame_diff, cv2.COLOR_BGR2GRAY)

    # Apply image smoothing
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Detect motion and changes
    _, thresh = cv2.threshold(blurred, 20, 255, cv2.THRESH_BINARY)

    # Find bounding rectangles
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        if cv2.contourArea(contour) > noise_threshold:
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame2, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the frame with bounding boxes using cv2_imshow
    # cv2_imshow(frame2)
    cv2.imshow('a', frame2)

    # Save the current frame as the previous frame
    frame1 = frame2

    # Read the next frame
    ret, frame2 = cap.read()

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Initialize the video capture
cap = cv2.VideoCapture('E:\_amirkabir/3.mp4')  # Use 0 for the default camera

# Create a background subtractor object
fgbg = cv2.createBackgroundSubtractorMOG2()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Apply background subtraction to the current frame
    fgmask = fgbg.apply(frame)

    # Perform morphological operations to clean up the mask
    fgmask = cv2.erode(fgmask, None, iterations=2)
    fgmask = cv2.dilate(fgmask, None, iterations=2)

    # Find contours in the mask
    contours, _ = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        if cv2.contourArea(contour) > 6000:  # Adjust the area threshold as needed
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the original frame with bounding boxes
    cv2.imshow("Motion Detection", frame)

    if cv2.waitKey(30) & 0xFF == 27:  # Press 'Esc' key to exit
        break

cap.release()
# cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from pydub import AudioSegment

# Initialize the video capture
cap = cv2.VideoCapture(r'E:\_amirkabir\3.mp4')  # Use 0 for the default camera

# Create a background subtractor object
fgbg = cv2.createBackgroundSubtractorMOG2()

# Initialize a list to store frames with bounding boxes
frames_with_boxes = []

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Apply background subtraction to the current frame
    fgmask = fgbg.apply(frame)

    # Perform morphological operations to clean up the mask
    fgmask = cv2.erode(fgmask, None, iterations=2)
    fgmask = cv2.dilate(fgmask, None, iterations=2)

    # Find contours in the mask
    contours, _ = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        if cv2.contourArea(contour) > 6000:  # Adjust the area threshold as needed
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Append the frame with bounding boxes to the list
    frames_with_boxes.append(frame)

    # Display the original frame with bounding boxes
    cv2.imshow("Motion Detection", frame)

    if cv2.waitKey(30) & 0xFF == 27:  # Press 'Esc' key to exit
        break

cap.release()
cv2.destroyAllWindows()

# Save the frames with bounding boxes as a video file
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter("output.avi", fourcc, 30, (640, 480))
for frame in frames_with_boxes:
    out.write(frame)
out.release()

# Extract audio from the saved video and save it as an MP3 file
video = cv2.VideoCapture("output.avi")
fps = int(video.get(5))
audio = AudioSegment.silent(duration=0)
while True:
    ret, audio_frame = video.read()
    if not ret:
        break
    audio += AudioSegment.from_wav("temp.wav")
audio.export("output.mp3", format="avi")
```

```python
import cv2
import numpy as np
from pydub import AudioSegment

# Initialize the video capture
cap = cv2.VideoCapture('E:\_amirkabir/3.mp4')  # Use 0 for the default camera

# Create a background subtractor object
fgbg = cv2.createBackgroundSubtractorMOG2()

# Initialize a list to store frames with bounding boxes
frames_with_boxes = []

# Get the original video dimensions and FPS
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(5))

# Define the codec and create a VideoWriter object with the same dimensions and FPS
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter("output.avi", fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Apply background subtraction to the current frame
    fgmask = fgbg.apply(frame)

    # Perform morphological operations to clean up the mask
    fgmask = cv2.erode(fgmask, None, iterations=2)
    fgmask = cv2.dilate(fgmask, None, iterations=2)

    # Find contours in the mask
    contours, _ = cv2.findContours(fgmask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        if cv2.contourArea(contour) > 1000:  # Adjust the area threshold as needed
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Append the frame with bounding boxes to the list
    frames_with_boxes.append(frame)

    # Save the frame to the video
    out.write(frame)

    # Display the original frame with bounding boxes
    cv2.imshow("Motion Detection", frame)

    if cv2.waitKey(30) & 0xFF == 27:  # Press 'Esc' key to exit
        break

cap.release()
out.release()
cv2.destroyAllWindows()

# Extract audio from the saved video and save it as an MP3 file
audio = AudioSegment.silent(duration=0)
for frame in frames_with_boxes:
    audio += AudioSegment.from_wav("temp.wav")
audio.export("output.mp3", format="mp3")
```
